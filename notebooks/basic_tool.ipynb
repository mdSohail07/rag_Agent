```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int , b:int):
    """it will reutrn the sum of two numbers

    Args:
    a : number one
    b : number two

    """
    return a + b
    
```

```python
@tool
def mul_numbers(a:int , b:int):
    """it will reutrn the mul of two numbers

    Args:
    a : number one
    b : number two

    """
    return a + b
```

```python
add_numbers.invoke({"a":12, "b":3})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm = ChatGoogleGenerativeAI(
    model="models/gemini-flash-latest",
    temperature=0.7
)
```

```python
agent = create_agent(
    model=llm,
    tools = [add_numbers],
    system_prompt = "you are a math teacher. you need to do calculation."

)
```

```python
res = agent.invoke({"messages":[{"role":"user", "content":"what is the sum of 1 + 1 * 3 ?"}]})
```
